```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_name = "dataset/p4-1_Male_20-29_170-179cm_Hand_held.txt"
```

**Read Data From Database**

```python
raw_data = []
accX = []
accY = []
accZ = []
accMagn = []
fr = open(file_name, 'r')
for line in fr.readlines():
    items = line.split(';')
    x = items[1]
    y = items[2]
    z = items[3]
    magn = items[4].split()[0]
    data_tuple = [float(x), float(y), float(z), float(magn)]
    raw_data.append(data_tuple)
    accX.append(float(x))
    accY.append(float(y))
    accZ.append(float(z))
    accMagn.append(float(magn))
fr.close()
print file_name.split('/')[1] + " *read over", len(raw_data)
```

**Raw Signal Figure (on time domain), ranging from 1000 to 1500**

```python
fig = plt.figure(figsize=(18, 5))
ax = fig.add_subplot(111, facecolor='#FFFFCC')

start = 1000
end = 1500

x_acc = np.asarray(accX[start:end])
xlist = range(start, end)
x = np.asarray(xlist)
ax.plot(x, x_acc, 'o-')

y_acc = np.asarray(accY[start:end])
xlist = range(start, end)
x = np.asarray(xlist)
ax.plot(x, y_acc, '*-')

z_acc = np.asarray(accZ[start:end])
xlist = range(start, end)
x = np.asarray(xlist)
ax.plot(x, z_acc, 's-')

plt.show()
```

**acc magn plot**

```python
fig = plt.figure(figsize=(18, 5))
ax = fig.add_subplot(111, facecolor='#FFFFCC')

start = 0
end = 6000

array_magn = np.asarray(accMagn[start:end])
x = np.asarray(range(start, end))
ax.plot(x, array_magn, 'o-')
plt.show()
```

```python
def maxWind(window):
    l = len(window)
    max = window[0]
    for i in range(1, l):
        if window[i] > max:
            max = window[i]
    return max

resInd = []
p = 4
wind_length = 15 # 150 ms
window = [0.0 for i in range(wind_length)]
time_out = 30 # 300 ms
timeout_cnt = 0
for i in range(len(accMagn)):
    if i == 0:
        for j in range(wind_length):
            window[j] = accMagn[i]
    else:
        for j in range(wind_length-1):
            # shift the window
            window[j] = window[j+1]
        window[wind_length - 1] = accMagn[i]
        
    if timeout_cnt > 0:
        timeout_cnt -= 1;
        continue

    max_wind = maxWind(window)
    if max_wind - window[wind_length-1] >= p:
        resInd.append(i)
        timeout_cnt = time_out

print len(resInd)
```

```python
fig = plt.figure(figsize=(18, 5))
ax = fig.add_subplot(111, facecolor='#FFFFCC')

start = 0
end = 6000

array_magn = np.asarray(accMagn[start:end])
x = np.asarray(range(start, end))
ax.plot(x, array_magn, 'v-')

mark_ind = []
for i in range(len(resInd)):
    if resInd[i] >= start and resInd[i] < end:
        mark_ind.append(resInd[i])
mark_list = []
for ind in mark_ind:
    mark_list.append(accMagn[ind])
    
marks = np.asarray(mark_list)
x = np.asarray(mark_ind)
ax.plot(x, marks, '^', markersize = 10)

plt.show()
```
